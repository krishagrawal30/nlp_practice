```python
import spacy
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
```

```python
# Load the English NLP model
nlp = spacy.load("en_core_web_sm")
```

```python
text = """
Artificial Intelligence is transforming the modern world. 
Narendra Modi visited the United States in 2023 and met Joe Biden in Washington.
Many organizations like Google and Microsoft are investing heavily in AI research.
"""
```

```python
# Process the text
doc = nlp(text)

print("Named Entities Detected:\n")

# Extract entities
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
# Tokenization
words = word_tokenize(text)

# POS Tagging
pos_tags = nltk.pos_tag(words)
```

```python
print("POS Tagged Words:\n")
print(pos_tags)
```

```python
allowed_tags = ['NN','NNS','JJ','VB','VBD','VBG','VBN','VBP','VBZ','RB']
```

```python
print("\nSynonyms and Antonyms:\n")
for word, tag in pos_tags:
    
    if tag in allowed_tags and word.isalpha():
        
        synonyms=set()
        antonyms=set()
        
        for syn in wordnet.synsets(word):
            for lemma in syn.lemmas():
                
                synonyms.add(lemma.name())
                
                if lemma.antonyms():
                    antonyms.add(lemma.antonyms()[0].name())
        
        if synonyms:
            print("Word:",word)
            print("Synonyms:",list(synonyms)[:5])
            print("Antonyms:",list(antonyms)[:5])
            print()
```
