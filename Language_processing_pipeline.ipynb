```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
nlp.pipeline
```

```python
nlp.pipe_names
```

```python
doc  = nlp('''On June 5, 2024, Elon Musk announced in New York that SpaceX secured $3 million from NASA, increasing revenue by 15%''')
```

```python
#pos_tag and lemma
for i in doc:
    print(i, " - ", i.pos_, " - ", i.lemma_)
    
```

```python
for i in doc.ents:
    print(i, " - ", i.label_, " - ", spacy.explain(i.label_))
```
