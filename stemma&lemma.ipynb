```python
import nltk
import spacy
```

```python
from nltk.stem import PorterStemmer
stm = PorterStemmer()
```

```python
words = [
    "relational",
    "conditional",
    "organizational",
    "activation",
    "happiness",
    "probability",
    "communities",
    "analysis",
    "generalization",
    "studies",
    "sensational",
    "adjustable",
    "saw",
    "left",
    "meeting",
    "better",
    "rose",
    "bears",
    "leaves",
    "lying",
    "bound",
    "found",
    "running",
    "cars"
]
for a in words:
    print(a, " - ", stm.stem(a))
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp(" ".join(words))

for token in doc:
    print(token, " -> ", token.lemma_)
```

```python
doc = nlp("Bro, listen brah, I told you dude and homie that this project isn’t easy")
for token in doc:
    print(token , " -> " , token.lemma_)
```

```python
#since both bro and brah are brother
ar = nlp.get_pipe("attribute_ruler")
ar.add([[{"TEXT":"Bro"}],[{"TEXT":"brah"}]], {"LEMMA" : "Brother"})
doc = nlp("Bro, listen brah, I told you dude and homie that this project isn’t easy")
for token in doc:
    print(token , " -> " , token.lemma_)
```
