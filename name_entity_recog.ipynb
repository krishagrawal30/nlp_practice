```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
nlp.pipe_names
```

```python
doc  = nlp('''On June 5, 2024, Elon Musk announced in New York that SpaceX secured $3 million from NASA, increasing revenue by 15%''')
```

```python
for i in doc.ents:
    print(i, " - ", i.label_, " - ", spacy.explain(i.label_))
```

```python
#use pip install spacy jupyter ipython notebook (if any error)
from spacy import displacy
displacy.render(doc, style = "ent")
```

```python
nlp.pipe_labels["ner"]
```

```python
#to make the ner classify SpaceX we use custom NER
doc[13:14]
```

```python
from spacy.tokens import Span
s = Span(doc, 13, 14, label="ORG")
doc.set_ents([s], default = "unmodified")
```

```python
for i in doc.ents:
    print(i, " - ", i.label_, " - ", spacy.explain(i.label_))
```

```python
from spacy import displacy
displacy.render(doc, style = "ent")
```
