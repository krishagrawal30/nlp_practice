```python
import spacy

nlp = spacy.load("en_core_web_sm")
print("Model loaded successfully ✅")
```

### Tokenization

```python
doc = nlp("Dr. Krish is also a scientist, researcher, Engineer, etc. apart from student. He loves to play basketball and chess.")
for i in doc.sents:
    print(i)
```

```python
for i in doc.sents:
    for w in i:
        print(w)
```

```python
import nltk
nltk.download("punkt_tab")
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
sent_tokenize("Dr. Krish is also a scientist, researcher, Engineer, etc. apart from student. He loves to play basketball and chess.")
```

```python
word_tokenize("Dr. Krish is also a scientist, researcher, Engineer, etc. apart from student. He loves to play basketball and chess.")
```

```python
nlp1 = spacy.blank("en")
```

```python
doc = nlp1("Dr. Krish is also a scientist, researcher, Engineer, etc. apart from student. He loves to play basketball and chess.")
for i in doc:
    print(i)
```

```python
nlp_hi = spacy.blank("hi")
hin_doc = nlp_hi("अरे भाई! इस स्पर्धा में ₹3000 का पुरस्कार रखा गया है। अब तो आप खेलेंगे न?")
for token in hin_doc:
    print(token, token.is_currency, token.like_num)
```

```python
 nlp_hi.add_pipe('sentencizer')
```

```python
hin_doc = nlp_hi("Dr. Krish is also a scientist, researcher, Engineer, etc. apart from student. He loves to play basketball and chess.")
for i in hin_doc.sents:
    print(i)
```
